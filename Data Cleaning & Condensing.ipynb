```python
# Functions to inspect data

# finds matches in columns of two data frames (independent of index)
def compare_cols(df1,df2,col_name):
    df1 = df1.reset_index()
    df2 = df2.reset_index()
    matches = []
    idx_df1 = []
    idx_df2 = []
    for i in range(len(df1)):
        for j in range(len(df2)):
            if (df1[col_name][i] == df2[col_name][j]):
                matches.append(df1[col_name][i])
                idx_df1.append(i)
                idx_df2.append(j)
    print('Lengths:', len(df1), ',', len(df2))
    print(len(matches), 'matches: ')
    print(matches)
    print('First dataframe indexes:', idx_df1)
    print('Second dataframe indexes:', idx_df2)
    print('\n')
    
# finds index of duplicates based on one column 
# necessary as we want to check more stats before we drop a duplicate
def get_duplicate_index(df, column):
    num_singles = 0
    for i in range(len(df)):
        if df[column].duplicated()[i] == True:
            print(i)
        else: 
            num_singles += 1
    if num_singles == len(df):
        print('No duplicates')
            
```
